# Gradient Descent 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([3,5,5,6,7,11,3])
y = 1 * X 
```

```python
X
```

```python
# Error function (Cost function)
def error_function(m, X, y):
    n = len(X)
    total_error = np.sum((y - m * X) ** 2)
    if m in {-2, -1, 0, 1, 2}:
        print(" slope: %d  error: %d" % (m, total_error / (2 * n)))
    return total_error / (2 * n)
```

```python
# Visualize the error function
slope = np.linspace(-2, 4, 100)
errors = [error_function(m, X, y) for m in slope]

plt.plot(slope, errors)
plt.title("Error Function")
plt.xlabel("Slope (m)")
plt.ylabel("Error")
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.show()
```
