## Normal Distribution Density 

```python
x=np.random.normal(scale=3,loc=20,size=100)
x
```

```python
# Calculate mean and standard deviation
mu = np.mean(x)
sigma = np.std(x)
```

```python
# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(x, bins='auto', kde=True, color='skyblue', edgecolor='black')
# Add vertical lines for μ, μ-σ, μ+σ
plt.axvline(mu, color='red', linestyle='--', label=f'Mean (μ) = {mu:.2f}')
plt.axvline(mu - sigma, color='green', linestyle='--', label=f'μ - σ = {mu - sigma:.2f}')
plt.axvline(mu + sigma, color='green', linestyle='--', label=f'μ + σ = {mu + sigma:.2f}')

# Add labels
plt.title('Histogram of Normal Distribution (μ=20, σ=3)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
